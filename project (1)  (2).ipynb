# Basic Dictionaries we will use  :

```python
days   = {'st':'Saturday','sn':'Sunday','mn':'Monday','tu':'Tuesday','wn':'Wensday','th':'Thursday','fr':'Friday','0':'all'}
cities = { 'c': 'chicago','n': 'new_york_city','w': 'washington' }
months = {'1':'january', '2':'february', '3':'march', '4':'april', '5':'may', '6':'june','7':'july','8':'august','9':'september','10':'october','11':'november','12':'december','0':'all'}
filtr  = {'1':'By Month','2':'By Day','3':'Both','0':'not at all'}
```

# get_filters() : method  which aim to Ask user to specify a city, month, and day to analyze and display certain data depending on his choice 


```python
def get_filters():
    """
    Asks user to specify a city, month, and day to analyze.

    Returns:
        (str) city - name of the city to analyze
        (str) month - name of the month to filter by, or "all" to apply no month filter
        (str) day - name of the day of week to filter by, or "all" to apply no day filter
        (str) fltr - which display user filtration choice

    """
    print('Hello! Let\'s explore some US bikeshare data!') 
    fltr,month,day='','',''
    while True :
        fltr =input ('do you want to filter data by : \n Enter (1) for by month \n Enter (2) for by day , \n Enter (3) for by both day and month  or\n Enter (0) for  not at all ').lower()
        if fltr not in filtr.keys():
            print("\n"+correct_choose_msg)
        else :
            break 
            

    # TO DO: get user input for city (chicago, new york city, washington). HINT: Use a while loop to handle invalid inputs
    while True :
        city =input (mesg_city).lower()
        if city not in cities.keys():
            print("\n"+correct_choose_msg)
        else :
            break 
    # TO DO: get user input for month (all, january, february, ... , june)
    if fltr =='1':
        day='none'
    elif fltr =='2' :
        month='none'
    elif fltr =='0':
        month='none'
        day='none'

    if month!='none':
        while True :
                    month = input (mesg_month).lower()
                    if month in months.keys() :
                        break 
                    else :
                        print("\n"+correct_choose_msg)

    # TO DO: get user input for day of week (all, monday, tuesday, ... sunday)
    if day!='none' :
        while True :
            day = input (mesg_day).lower()
            if day in days.keys() :
                break 
            else :
                print("\n"+correct_choose_msg)
                
    print('-'*40)
    return city, month, day, fltr
#will  be a tuple of each filter user choose of city , month , day respectively 
#city may be (c,n,w)
#month may be (0-12),0 refer to all 
#day may be (0-(st,sn,mn,tu,wn,th,fr)),0 refer to all 
```

# Load_data () : method aim to load subset of relevant data from original dataset to display it for the user after taking  his filtration for (city, month, day) which we ask him before, method take them as an argumens

```python
def load_data(city, month, day , fltr):
    """
    Loads data for the specified city and filters by month and day if applicable.

    Args:
        (str) city - name of the city to analyze
        (str) month - name of the month to filter by, or "all" to apply no month filter
        (str) day - name of the day of week to filter by, or "all" to apply no day filter
    Returns:
        df - Pandas DataFrame containing city data filtered by month and day
    """

    # load data file into a dataframe
    if (city=='c'):
        df=chicago 
    elif (city=='n'):
        df=new_york
    elif (city=='w'):
        df=washington

    # convert the Start Time column to datetime 
    df['Start Time'] = pd.to_datetime(df['Start Time']) #inplace conversion from string to datetime  
    # extract month and day of week from Start Time to create new columns
    df['month'] = df['Start Time'].dt.month
    df['day_of_week'] = df['Start Time'].dt.day_name()
    df['start_hour']=df['Start Time'].dt.hour
    # filter by month if applicable
    
    if fltr =='1'or fltr=='3':
        if month != '0'and month != 'none' :
        # use the index of the months list to get the corresponding int
        ##monthsss = ['january', 'february', 'march', 'april', 'may', 'june']
        #m = monthsss.index(month) + 1
        # filter by month to create the new dataframe
            df = df[df['month'] == int(month)]
    # filter by day of week if applicable
    if fltr =='2'or fltr=='3':
        if day != '0' and day!='none':  #day may be (st,sn,mn,tu,wn,th,fr) as      
        # filter by day of week to create the new dataframe
            d  = days[day] # get croessponding day name from dictionary days as day is the key 
            df = df[df['day_of_week'] == d]

    return df
```

# time_stats() : method which Displays statistics on the most frequent times of travel


```python
def time_stats(df,flt):

    print('\nCalculating The Most Frequent Times of Travel...\n')
    start_time = time.time()
    
    # TO DO: display the most common month
    M=df['month'].mode()[0]
    conv=calendar.month_name[M]
    print('most common month :', conv )
    print('count :', df['month'].value_counts()[M])
    print('filter :' ,filtr[flt])  

    # TO DO: display the most common day of week
    D=df['day_of_week'].mode()[0]
    print('most common day :', D )
    print('count :',df['day_of_week'].value_counts()[ D ])
    print('filter :' ,filtr[flt])  

    # TO DO: display the most common start hour
    hr=df['start_hour'].mode()[0]
    print('most common start hour : ',hr)
    print('count :' , df['start_hour'].value_counts()[ hr ])
    print('filter :' ,filtr[flt])  

    print("\nThis took %s seconds." % (time.time() - start_time))
    print('-'*40)
```

# station_stats() methods which Displays statistics on the most popular stations and trip.


```python
def station_stats(df,flt):

    print('\nCalculating The Most Popular Stations and Trip...\n')
    start_time = time.time()

    # TO DO: display most commonly used start station
    st=df['Start Station'].mode()[0]
    print(' most commonly used start station is {} '.format(st))
    print('count :', df['Start Station'].value_counts()[st])
    print('filter :', filtr[flt])

    # TO DO: display most commonly used end station
    end=df['End Station'].mode()[0]
    print('most commonly used end station is {}'.format(end))
    print('count :' , df['End Station'].value_counts()[ end ])
    print('filter :' , filtr[flt])    

    # TO DO: display most frequent combination of start station and end station trip
    df['trip']= df['Start Station']+","+df['End Station']  
    most_frequent_trip =df['trip'].mode()[0]
    print('most frequent trip  {} '.format(most_frequent_trip))
    print('count :', df['trip'].value_counts()[most_frequent_trip])
    print('filter :' ,filtr[flt])  

    print("\nThis took %s seconds." % (time.time() - start_time))
    print('-'*40)
```

# trip_duration_stats() method which Displays statistics on the total and average trip duration.

```python
def trip_duration_stats(df,flt):

    print('\nCalculating Trip Duration...\n')
    start_time = time.time()

    # TO DO: display total travel time
    print('total travel time is {}'.format( df['Trip Duration'].sum()))

    # TO DO: display mean travel time
    print('mean travel time is {}'.format( df['Trip Duration'].mean()))


    print("\nThis took %s seconds." % (time.time() - start_time))
    print('-'*40)
```
